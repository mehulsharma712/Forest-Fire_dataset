```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt    
import seaborn as sns  
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.info()
```

Data cleaning

```python
##Missing values    
dataset[dataset.isnull().any(axis=1)]
```

the dataset is converted into two sets based on Region from 122th index, can make a new column based on the region                

Add new column with region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
##Removing the null values    
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
##Removing the 122th index    
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

There are soo much unnecessary spaces in the column names

```python
##Fix  spaces in column names    
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

Changing the required columns to int

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

Changing the other columns to float data type

```python
objects=[features for features in df.columns if df[features].dtype==object]
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv',index=False)
```
